```python
products = ['t-shirt', 'mug', 'hat', 'book', 'keychain']
```

```python
def initialize_inventory(products):
    inventory = {}
    for product in products:
         quantity = int(input(f'Enter the quantity for {product}: '))
         inventory[product] = quantity
    return inventory
```

```python
inventory = initialize_inventory(products)
```

```python
inventory
```

```python
customer_orders = set()
```

```python
def get_customer_orders():
    customer_orders = set()
    for i in range(3):
        product = input('Enter a product: ')
        customer_orders.add(product)
    return customer_orders
```

```python
orders = get_customer_orders()
```

```python
orders
```

```python
def update_inventory(customer_orders, inventory):
    for product in customer_orders:
        inventory[product] -= 1
    return inventory
```

```python
inventory = update_inventory(orders, inventory)
```

```python
inventory
```

```python
def calculate_order_statistics(customer_orders, products):
    total_products_ordered = len(customer_orders)
    percentage_of_unique_products_ordered = (total_products_ordered / len(products)) * 100
    return total_products_ordered, percentage_of_unique_products_ordered
```

```python
order_statistics = calculate_order_statistics(orders, products)
```

```python
order_statistics
```

```python
def print_order_statistics(order_statistics):
    total_products_ordered, percentage_of_unique_products_ordered = order_statistics

    print('Order Statistics:')
    print('Total Products Ordered:', total_products_ordered)
    print('Percentage of unique products Ordered:', percentage_of_unique_products_ordered)
```

```python
print_order_statistics(order_statistics)
```

```python
def print_update_inventory(inventory):
    for product in inventory:
        print(product,":", inventory[product])
```

```python
print_update_inventory(inventory)
```

```python
products = ['t-shirt', 'mug', 'hat', 'book', 'keychain']

inventory = initialize_inventory(products)

orders = get_customer_orders()

inventory = update_inventory(orders, inventory)

order_statistics = calculate_order_statistics(orders, products)

print_order_statistics(order_statistics)

print_update_inventory(inventory)
```
