```python
def initialize_inventory(products):
    inventory = {}
    for product in products:
        quantity = int(input(f"Enter the quantity of {product} in inventory: "))
        inventory[product] = quantity
    return inventory
```

```python
initialize_inventory(["t-shirt", "mug"])
```

```python
inventory = initialize_inventory(["t-shirt", "mug"])
```

```python
inventory
```

```python

def get_customer_orders():
    customer_orders = set()
    continue_entering = 'yes'
    while continue_entering.lower() == 'yes':
        product_order = input('Enter the product you want to order: ')
        customer_orders.add(product_order)

        continue_entering = input('Do you want to add another product? (yes/no)').lower()

    return customer_orders
```

```python
customer_orders = get_customer_orders()
```

```python
customer_orders 
```

```python
def update_inventory(customer_orders, inventory):
    for item in customer_orders:
        if item in inventory:
            inventory[item] -= 1
    
    return inventory
```

```python
inventory = update_inventory(customer_orders, inventory)
```

```python
inventory
```

```python

def calculate_order_statistics(customer_orders, products):
    total_set = len(customer_orders)
    percentage_ordered = (total_set / len(products)) * 100
    order_status = (total_set, percentage_ordered)
    return order_status
```

```python
products = ["t-shirt", "mug"]
order_statistics = calculate_order_statistics(customer_orders, products)

```

```python
order_statistics
```

```python
def print_order_statistics(order_statistics):
    print(order_statistics)
    
```

```python
order_statistics
```

```python
def print_updated_inventory(inventory):
    for product, quantity in inventory.items():
        print(product, quantity)
```

```python
inventory = print_updated_inventory(inventory)
```

```python
inventory
```

```python
products = ["t-shirt", "mug"]

inventory = initialize_inventory(products)
customer_orders = get_customer_orders()
inventory = update_inventory(customer_orders, inventory)
order_statistics = calculate_order_statistics(customer_orders, products)
print_order_statistics(order_statistics)
print_updated_inventory(inventory)
```
